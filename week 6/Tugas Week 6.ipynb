```python
for i in range(1, 10):
    for y in range(1, 10+1):
      print(i * y, end = " ")
    print()
```

```python
names = ['Yasa' , 'Athalla', 'Jaya']
for name in names:
    count = 0
    while count < 5 :
        print(name, end = ' ')
        count = count + 1
    print()
```

```python
for i in range(4):
    for j in range(4):
        if j == i:
            break
        print(i, j)
```

```python
pertama = [2, 4, 6]
kedua = [1, 3, 5]
for  i in pertama:
    for j in kedua:
        if i == j :
            continue
        print(i, '*', j, '= ', i * j)
```

```python
n = 2
while n <= 100:
    x_sum = 0
    for i in range(1, n):
        if n % i == 0:
            x_sum += i
    if x_sum == n:
        print('Perfect number:', n)
    n += 1
```

```python
tinggi = int(input("Masukkan tinggi segitiga: "))
for i in range(tinggi):
    for j in range(tinggi - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

```python
dasar = 1
total = 0
angka_baris = int(input("Masukkan angka: "))

for i in range(angka_baris):
    if i == 0:
        dasar = 1
    else:
        dasar = dasar * (angka_baris - i) // i
    print(dasar, end=" ")
    total += dasar

print(f"\nJumlah Angka dalam baris {angka_baris} adalah: {total}")
```
