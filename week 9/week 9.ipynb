```python
#Return the third, fourth, and fifth item:
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango", "mango"]
print(thislist[2:5])
#return item tapi tanpa index ke 4 atau  kiwi
print(thislist[:4])
```

```python
#return item dimulai dari index 2 atau cherry sampe ujung
print(thislist[2:])
```

```python
#return item dari -4 = kiwi sampai -1 tpi tanpa -1
print(thislist[-4:-1])
```
