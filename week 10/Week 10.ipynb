```python
n = int(input("Masukkan angka: "))
def recursion(n):
    if n > 1 :
      return n * recursion(n - 1)
    return 1

rekursi = recursion(n)
print(f'{n}! = {rekursi}')
print()
```

```python
def lingkaran() :
    print("Hitung luas lingkaran")
    phi = 3.14
    jari = float(input("Masukkan jari-jari:"))**2
    lingkaran = phi * jari
    print(f'Luas lingkaran adalah {lingkaran}\n')

def persegi() :
    print("Hitung luas persegi")
    sisi = float(input("Masukkan sisi: "))
    persegi =  sisi * sisi
    print(f'Luas persegi adalah {persegi}\n')

def segitiga() :
    print("Hitung luas segitiga")
    alas = float(input("Masukkan alas: "))
    tinggi = float(input("Masukkan tinggi: "))
    segitiga = 0.5 * alas * tinggi
    print(f'Luas segitiga adalah {segitiga}\n')

while True:
    user = int(input("Pilih rumus yang ingin digunakan: \n\n1.Luas Lingkaran\n2.Luas Persegi\n3.Luas Segitiga\n0.Keluar program\n\nPilih rumus yang tersedia: "))
    if(user == 1):
        lingkaran()
    elif(user == 2):
        persegi()
    elif(user == 3):
        segitiga()
    elif(user == 0):
        break
    else:
        print("Pilihan tidak tersedia")



```
